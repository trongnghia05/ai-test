```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def init_point(point):
    X = []
    Y = []
    for i in range(point):
        x = np.random.randint(low = 5 , high = 100 + i)
        y = np.random.randint(low = x +50, high = x + 100)
        X.append([x, 1])
        Y.append(y)
    X = np.array(X)
    Y = np.array(Y)
    return X, Y.reshape(-1,1)

def fit(X, Y, W, epoch = 20, lr = 0.00001):
    for i in range(epoch):
        Y_pred = np.dot(X,W)
        loss = np.sum((Y_pred - Y)**2)/(2*point)
        x_w_y = np.dot(X, W) - Y
        d_w = np.dot(X.T, x_w_y)
        W = W - lr*d_w/point
        if (i%10000==0):
            print("Epoch {}, loss = {}".format(i, loss))
    return W
```

```python
point = 100
# khởi tạo các cặp điểm
X, Y = init_point(point)

#khởi tạo W
W = np.random.rand(2,1)

#fit
W = fit(X, Y, W, epoch = 100000, lr = 0.00006)
print("W: {}".format(W))

#draw
fig, ax = plt.subplots(1,figsize=(10,10))
ax.scatter(X[:,0],Y)
Y_pred = np.dot(X, W).reshape(-1)
ax.plot(X[:,0],  Y_pred, 'r')
```
